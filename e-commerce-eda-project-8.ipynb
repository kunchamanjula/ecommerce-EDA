# 1.provide the list of different arthemetic values based on the values of MRP of different products in dataset using group aggregation?

```python
 cost=ecommerce.groupby(by="MRP")["Discount"].agg(func=["mean", "median", "max", "min"]).sort_values(by="mean", ascending=False)
cost#using groupby aggratation to know the values of min max,median in MRP  and discount of  products
```

# 2.create different categorises based on sizes of products by defining new function?

```python
#categorzing the sizes of products 
def product_sizes(value):
    
    if value=="Size:Large,Medium,Small,X-Large,X-Small":
        return "All sizes of clothes"
    elif value=="Size:34,36,38,40" :
        return "bottom sizes"
    elif value =="Size:Large,X-Large,XX-Large":
        return "large size"
    elif value== 'Size:4,4.5,8.5':
        return "shoe size"
    elif value =="Size:XL,X-Small":
        return "small sizes"
    else:
        return "all sizes available"

# now create the categories according to the values
ecommerce["product_sizes"] = ecommerce["Sizes"].apply(lambda x:product_sizes(x)).copy( )
```

```python
ecommerce.head() # new column has been added to dataset on the name of product sizes
```

# 3.plot the graph which shows the sales of category based on their discounts wrt product sizes given in data?

```python

sns.catplot(data=ecommerce, x="Category", y="Discount", col="product_sizes", kind='box', col_wrap=2, palette="magma")

plt.show()
```

# 4.plot the pairplot which shows the disturbation of various variables present in dataset?

```python
sns.pairplot(ecommerce,vars=["MRP","SellPrice","Discount"],hue="product_sizes",palette="spring")
plt.show()
```

# 5..plot the piechart based on the categories of product based on mrp in data?

```python
ecommerce["Category"].value_counts().plot.pie(title="Product categories", autopct='%.2f', 
                                      figsize=(6, 6))# we can see that there are very less people who are buys fragance via ecommerce
plt.show()
```

# 6.find out which product sizes  are getting more and less purchased through ecommerce?

```python
ecommerce.groupby(by="product_sizes")["SellPrice"].mean().reset_index().sort_values(by="SellPrice", ascending=False).plot.line(title="PRODUCTS sales based on their Sizes." ,figsize=(10, 5), x="product_sizes",color="purple")
plt.show() #using chain method first done group aggreagation and then using line plot to get plotting
```

* small sizes are more getting purchased than other sizes on ecommerce

# 7.plot the graph which shows the disturbution of sales of products based on their catergorie and sizes via ecommerce?

```python
plt.figure(figsize=(12,5))
sns.boxenplot(y="Discount", hue="product_sizes", x="Category",data=ecommerce)
plt.title('Product sales based on categorie and sizes ', fontsize=16)
plt.xticks(rotation=45)

plt.show()
```

# 8. plot the graph to show the sell prices of product with different sizes in ecommerce?

```python
plt.figure(figsize=(12,5))
sns.violinplot(x="product_sizes", y="SellPrice", data=ecommerce)
plt.title('Prices of different brands based on their sizes ', fontsize=16)
plt.xticks(rotation=45)

plt.show()
```
